# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jypyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = np.random.randint(10,19, size=(2, 3) )
y = np.random.randint(10,20 ,size=(2, 3) )

# 2. Setup plot
fig, ax = plt.subplots(figsize = (5, 5))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Simple Plot", 
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10, 
                    "Peanut butter": 8, 
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title= "This is butter store", 
      ylabel= "Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplots option 1
fig, ((ax1, ax2),(ax3, ax4))= plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize= (10, 5),)
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax  = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### which one should you use? (pyplot vs marplotlib OO method?)
* when plotting something quickly, okay to use the pyplot method
* when plotting something more advanced, use the OO method


```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind = "scatter", 
            x="age", 
            y="chol",
            c="target");
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = "scatter", 
            x="age", 
            y="chol",
            c="target",
            ax=ax);
# ax.set_xlim([45, 100]);
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
#Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol");

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50.head()
```

```python
## sub plot of cholestrol, age and thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                              sharex=True);
# add data to ax0

scatter = ax0.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"])
# customize ax0
ax0.set(title="Hear Disease and Cholestrol Levels", 
       ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle = "--");


## add data ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
ax1.set(title="Age and Max Heart Rate", 
       xlabel="Age",
       ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), 
          linestyle="--");

# add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Use the style
plt.style.use("seaborn-v0_8-whitegrid")
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                    cmap="plasma"); #changes the color
#Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol");

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Custommizing Y and X axis limitations

## sub plot of cholestrol, age and thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                              sharex=True);
# add data to ax0

scatter = ax0.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"], 
                     cmap="winter")
# customize ax0
ax0.set(title="Hear Disease and Cholestrol Levels", 
       ylabel="Cholesterol");

# set x axis limit
ax0.set_xlim([50, 80])

ax0.legend(*scatter.legend_elements(), title="Target")

#Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle = "--");


## add data ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")
ax1.set(title="Age and Max Heart Rate", 
       xlabel="Age",
       ylabel="Max Heart Rate")


# set x axis limit
ax1.set_xlim([50, 80])


# set y axis limit
ax1.set_ylim([60, 200])

ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), 
          linestyle="--");

# add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
